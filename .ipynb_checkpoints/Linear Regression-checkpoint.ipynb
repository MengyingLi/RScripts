### Some important terms

* For simple linear regression, the coefficient is cov(x, y)/var(x)
* Correlation: indicate how closely their relationship follows a straight line (linear relationship). Rule of a thumb: 0.1-0.3 means weak, 0.3-0.5 means moderate and 0.5+ is strong. If the data generation involves human beings, 0.5+ is considered extremely high but when it is generated by mechanical process, a correlation of 0.5 might be weak
* Sample size: depends on how much power you want to get 

* Why skewness in predictor/outcome variable is bad for OLS

    while having a highly skewed dependent variable does not violate an assumption, it may make OLS regression rather inapporpriate. OLS regression models the mean and the mean is (usually) not a good measure of central tendency in a skewed distribution.  The median is often better and it can be modeled with quantile regression. In addition, when the DV is highly skewed the interest may be in modeling the tails of the distribution - this can also be done with quantile regression.

```r
reg <- function(y, x) {   
    x <- as.matrix(x)   
    x <- cbind(Intercept = 1, x)  
    b <- solve(t(x) %*% x) %*% t(x) %*% y   # solve is to get the inverse of a matrix; %*% multiply two matrices
    colnames(b) <- "estimate"   
    print(b) }
```

### Assumptions of linear regression

* Output is a weighted linear function of feature variables
* For fixed values of the feature variables, the output is normally distributed with a constant variance (homoscedasticity). Seldom tested in practice as as long the group size is equal and the sample is large, does not matter plus we can use Walch t to correct for the heteroskedasticity (Constant variance of errors versus the predictions and independent variables) 
* Independence of the residuals (for example if we take our training data and double the value of x and y and put them back to the data, we would have pairs of observations whose error terms will depend on each other as a result or the example where the observations are related in some way by an unmeasured variable) 
* No perfect collinearity between variables
* Non-zero variance

#### Multicolinearity

* As collinearity increases so do the standard errors of the b coefficients. If
you think back to what the standard error represents, then big standard errors for b
coefficients means that these bs are more variable across samples. Therefore, it means that
the b coefficient in our sample is less likely to represent the population. Crudely put,
multicollinearity means that the b-values are less trustworthy. Don’t lend them money and
don’t let them go for dinner with your boy-or girlfriend. Of course if the bs are variable from
sample to sample then the resulting predictor equations will be unstable across samples too.
    
* It limits the size of R sqaured. The variance of the variable accounts for the whole model is the same variance accounted for by another variable. The unique variance gone.

* Hard to assess the importance of predictors

### Transform the variables if needed for linearity assumption. 

* For linear and logistic regression, for example, you ideally want to make sure that the relationship between input variables and output variables is approximately linear, that the input variables are approximately normal in distribution, and that the output variable is constant variance (that is, the variance of the output variable is independent of the input variables)

* Residual section to see how far off we are
* For each estimated regression coefficient, the p-value , denoted Pr(>|t|) , provides an estimate of the probability that the true coefficient is zero given the value of the estimates. So the t can be interpreted as how many standard deviations away this coefficient is from 0.  it
is easiest to conceptualize the t-tests as measures of whether the predictor is making a significant
contribution to the model.
* Multiple R2 is the square of the correlation between the observed values of Y and the values of Y predicted by the multiple regression model
* The adjusted R-squared value corrects R-squared by penalizing models with a large number of independent variables. It is useful for comparing the performance of models with different numbers of explanatory variables.
* F-test tells us whether the model explained some variance compared to the base model. It is constructed by mean of SSM divided by mean of SSE. how much the model has improved the prediction of the outcome compared to the level of inaccuracy of the model. A good model should have a large F-ratio

* the
adjusted R sqaured tells us how much variance in Y would be accounted for if the model had been
derived from the population from which the sample was taken

<p ><font color ="red"> For a high p-value, we can't say that there is no linear relationship between the two variables, we can only say that in the presence of other variables, the feature does not provide any new info about the output. Significance test are not useful for high dimensionality problems as it is easy for us to get a very high numbers of wrong coefficients by 95% CI definition. </font> </p>

Strictly speaking, regression modeling makes some strong assumptions about the data. These assumptions are not as important for numeric forecasting as the model's worth is not based on whether it truly captures the underlying process. We only care about the accuracy of the prediction. 

### How to select the variables for the model?

* Stepwise selection. Not recommended. they assess the fit of a variable based on the other
variables that were in the model. Backward selection is more preferred as it is less likely to make type 2 error which means which occur when a predictor has an effect but only when another variable is held constant. Use AIC/BIC to assess the fit

* All subset regression. Use Mallows' Cp to access the fit.  Using Mallows' Cp to compare regression models is valid only when you start with the same complete set of predictors.  A small Mallows' Cp value indicates that the model is relatively precise (has small variance) in estimating the true regression coefficients and predicting future responses. 

### Finding outliers: Residual Analysis

* Why? Residuals reveal variaous aspects of our model, from violated assumptions and the quality of the fit to other problems such as outliers

* We need to compare the residuals with the mean of the output variable to see whether the residuals are large or not

#### Use Q-Q plot

```r
qqnorm(cars_model2$residuals)
qqline(cars_model2$residuals)
# Compare the distributions by comparing the values at their quantiles. 
# each interval has the same probability. If two distributions are the same, then the graph should be a plot of the line y = x
```

#### Use Residual plot

* Do a scatterplot of the residuals against each predictor in the model. If these scatterplots show a pattern, then the relationship may be nonlinear and the model will need to be modified accordingly
* Also do a scatterplot against the predictors which are not involved in the model before to see whether this one should be further included.

* Confirm constant variance to see whether the value of the variance has nothing to do with the value of y hat
* To see whether there is a sort of pattern in the residuals. If a pattern is observed, it may be an indication that 
the underlying model is nonlinear in terms of the feaures or we are missing features. One way to look for new features is to look for the features which are correlated with residuals. A plot of the residuals against the fitted values should show no pattern. If a pattern is observed, there may be "heteroscedasticity" in the errors. That is, the variance of the residuals may not be constant. To overcome this problem, a transformation of the forecast variable (such as a logarithm or square root) may be required.

```r
plot( cars_model2$fitted.values, cars_model2$residuals)
```

(1) standardized residuals with an absolute value greater than 3.29 (we can use 3 as an approximation) are cause for concern
because in an average sample a value this high is unlikely to happen by chance; (2) if more than1% of our sample cases have standardized residuals with an absolute value greater than 2.58 (we
usually just say 2.5) there is evidence that the level of error within our model is unacceptable (the
model is a fairly poor fit of the sample data); and (3) if more than 5% of cases have standardized
residuals with an absolute value greater than 1.96 (we can use 2 for convenience) then there is
also evidence that the model is a poor representation of the actual data.

### Outliers 

* An observation whose predicted value is very far from its actual value. Due to the squaring of residuals, the outliers tend to have a significant impact on the RSS, giving us a sense the model is not a good fit.

* Outliers can be due to measurement errors and detecting them may be important, or outliers may simply be the result of not having the right feautes or building the wrong kind of the model

* Use Residual Analysis

### High leverage points

* Observations that have an extreme value for at least one of their features and thus lie far away from most of the other observartions. Cook's distance is used to detect both outliers and high leverage points

* adjusted predicted value for a case when that case is excluded from the analysis. If the model is stable, the difference between the adjusted predicted value and the original predicted value is known as DFFit. When this residual is divided by the standard error, it gives a standardized value known as the studentized residual. They don't provide the direction of the influence.

* Cook's distance: a measure of the overall influence of a case on the model. Value greather than 1 may be cause for concern.
* Leverage/hat values: gauge the influence of the observed value of the outcome variable over the predicted values. The average influence is (k+1)/n.
