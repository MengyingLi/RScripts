## Analytical Distribution

### P-value(Default lower tail = true) Cumulative Function

* Figure out the corresponding pvalue to a specific value

```r
pnorm(0.1, mean =2, sd = 1, lower.tail = TRUE) #lower tail meaning less than the specified q (the first parameter)
```

### Percentile/Quantile

* Figure out the corresponding percentile to a percentile rank (can be upper or lower tail)

```r
qnorm(0.1, mean = 2, sd = 1, lower.tail = FALSE) #means the right tail has 0.1, what is the cutoff value corresponding to that tail
```

### Density

```r
dnorm(0.1, mean = 2, sd = 1) #Think it as the derivative of the cdf
```

```r
set.seed(3000)
xseq<-seq(-4,4,.01)
densities<-dnorm(xseq, 0,1)
cumulative<-pnorm(xseq, 0, 1)
randomdeviates<-rnorm(1000,0,1)
 
par(mfrow=c(1,3), mar=c(3,4,4,2))

plot(xseq, densities, col="darkgreen",xlab="", ylab="Density", type="l",lwd=2, cex=2, main="PDF of Standard Normal", cex.axis=.8)

plot(xseq, cumulative, col="darkorange", xlab="", ylab="Cumulative Probability",type="l",lwd=2, cex=2, main="CDF of Standard Normal", cex.axis=.8)

hist(randomdeviates, main="Random draws from Std Normal", cex.axis=.8, xlim=c(-4,4))
```

### Exponential Distribution

* When we look at a series of events and measure the times between events, caled interarrival times. If the events are equally likely to occur at any time, the distribution of interarrival times tends to look like an exponential distribution

```r
set.seed(3000)
xseq<-seq(2 ,4,.01)
densities<-dexp(xseq, rate =1)
cumulative<-pexp(xseq, rate = 1)
randomdeviates<-rexp(1000,rate = 1)
 
par(mfrow=c(1,3), mar=c(3,4,4,5))

plot(xseq, densities, col="darkgreen",xlab="", ylab="Density", type="l",lwd=2, cex=2, main="PDF of Exponential", cex.axis=.8)

plot(xseq, cumulative, col="darkorange", xlab="", ylab="Cumulative Probability",type="l",lwd=2, cex=2, main="CDF of Exponential Distribution", cex.axis=.8)

hist(randomdeviates, main="Random draws from Exponential Distribution", cex.axis=.8, xlim=c(-4,4))
```
